```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv(r"C:\Users\Maple\Downloads\energy1.csv")

df_clean = df[(df['TOTAL THERMS'] <= 120000) & (df['TOTAL KWH'] <= 550000)]
#A dropdown filter
building_types = ['Commercial', 'Residential']
#single = alt.selection_single()

chart = alt.Chart(df_clean).mark_circle(size=100).encode(
    x='TOTAL KWH:Q',
    y='TOTAL THERMS:Q',
    tooltip=[alt.Tooltip('TOTAL KWH', title = 'Total Kilowatts'), 
             alt.Tooltip('TOTAL THERMS', title = 'Total Therms'),
             alt.Tooltip('BUILDING TYPE', title = 'Building Type')]
).properties(
    width=800,
    height=400,
    title='Total Kilowatt & Therms Usage based on Building Type'
).interactive(bind_y=False)

dropdown = alt.binding_select(options=building_types, name="Type of Building")
select = alt.selection_point(fields=['BUILDING TYPE'], bind=dropdown)

filter = chart.add_params(
    select
).transform_filter(
    select
)

filter
```

```python
#saving my file to jekyll 
myJekyllDir = "C:\\Users\\Maple\\trane07.github.io\\assets\\json\\"

filter.properties(width='container').save(myJekyllDir+'proj3interact.json')
```
